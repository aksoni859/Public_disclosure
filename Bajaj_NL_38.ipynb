```python
# Import Libraries
import camelot
import os
import numpy as np
import pandas as pd
```

```python
# Fill data as per requirement
company = 'Bajaj_Allianz'
financial_yr = '2020_21'
quarter = 'Q4'
```

```python
LOB = ['Fire', 'Marine Cargo', 'Marine Others', 'Marine Total', 'Motor OD', 'Motor TP', 'Motor Total', 'Liability', 'Engineering', 'PA', 'Health Insurance', 'Crop Insurance', 'Others']
year = np.full(shape=len(LOB),fill_value=financial_yr)
qtr = np.full(shape=len(LOB),fill_value=quarter)
cmpny = np.full(shape=len(LOB),fill_value=company.replace('_',' '))
```

```python
dir_path = r"C:\Users\n179950\Munich Re\MR India NL - Documents\Agro\6. Miscellaneous\Individual's Folders\Anil\project_public_disclosure\PDF"
file = os.path.join(dir_path,company,financial_yr,quarter,'NL-38.pdf')
print(file)
```

```python
odir_path = r"C:\Users\n179950\Munich Re\MR India NL - Documents\Agro\6. Miscellaneous\Individual's Folders\Anil\project_public_disclosure\CSV"
ofile = os.path.join(odir_path,company,financial_yr,quarter,'NL-38.csv')
print(ofile)
```

```python
def chk_blank(value):
    if value =='-' and len(value)==1:
        return None
    elif len(value)==0:
        return None
    elif '(' in value:
        return int(value.replace('-','').replace('(','').replace(')','').replace(' ','')) *-1
    else:
        return int(value.replace('-','').replace(' ',''))
    
def ifnone(value):
    if value == None:
        return 0
    else:
        return value
```

```python
tables = camelot.read_pdf(file)#,pages="all")
print("Total tables extracted:", tables.n)
```

```python
##### Read Data from NL38 form

# Number of Policies
No_of_policies = []

No_of_policies_business_Fire = chk_blank(tables[0].df[2][2].replace(',',''))
No_of_policies_business_Marine_cargo = None
No_of_policies_business_Marine_others = None
No_of_policies_business_Marine_total = chk_blank(tables[0].df[2][3].replace(',',''))
No_of_policies_business_Motor_OD = chk_blank(tables[0].df[2][5].replace(',',''))
No_of_policies_business_Motor_TP = chk_blank(tables[0].df[2][4].replace(',',''))
No_of_policies_business_Motor_Total = No_of_policies_business_Motor_OD + No_of_policies_business_Motor_TP
No_of_policies_business_worksmens = None
No_of_policies_business_Liability = chk_blank(tables[0].df[2][7].replace(',',''))
No_of_policies_business_Engineering = chk_blank(tables[0].df[2][6].replace(',',''))
No_of_policies_business_aviation = chk_blank(tables[0].df[2][12].replace(',',''))
No_of_policies_business_PA = chk_blank(tables[0].df[2][8].replace(',',''))
No_of_policies_business_Health_Insurance = chk_blank(tables[0].df[2][9].replace(',',''))
No_of_policies_business_Credit_Insurance = chk_blank(tables[0].df[2][13].replace(',',''))
No_of_policies_business_Crop_Insurance = chk_blank(tables[0].df[2][11].replace(',',''))
No_of_policies_business_Others = chk_blank(tables[0].df[2][14].replace(',',''))


### Add others all LoB
No_of_policies_business_Others = ifnone(No_of_policies_business_Others) + ifnone(No_of_policies_business_worksmens) \
                               + ifnone(No_of_policies_business_aviation) + ifnone(No_of_policies_business_Credit_Insurance)

No_of_policies.append(No_of_policies_business_Fire)
No_of_policies.append(No_of_policies_business_Marine_cargo)
No_of_policies.append(No_of_policies_business_Marine_others)
No_of_policies.append(No_of_policies_business_Marine_total)
No_of_policies.append(No_of_policies_business_Motor_OD)
No_of_policies.append(No_of_policies_business_Motor_TP)
No_of_policies.append(No_of_policies_business_Motor_Total)
No_of_policies.append(No_of_policies_business_Liability)
No_of_policies.append(No_of_policies_business_Engineering)
No_of_policies.append(No_of_policies_business_PA)
No_of_policies.append(No_of_policies_business_Health_Insurance)
No_of_policies.append(No_of_policies_business_Crop_Insurance)
No_of_policies.append(No_of_policies_business_Others)
```

```python
data = {'Year': year,
         'Qtr': qtr,
         'Company':cmpny,
         'LOB':LOB,
         'Number of Policies': No_of_policies}
```

```python
df_data = pd.DataFrame(data)
df_data.to_csv(ofile,index=False)
```
