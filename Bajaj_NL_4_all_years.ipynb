```python
##### Extract information from all years

# Import Libraries
import camelot
import os
import numpy as np
import pandas as pd

def chk_blank(value):
    if value =='-' and len(value)==1:
        return None
    elif len(value)==0:
        return None
    elif '(' in value:
        return int(value.replace('-','').replace('(','').replace(')','').replace(' ','')) *-1
    else:
        return int(value.replace('-','').replace(' ',''))
    
def ifnone(value):
    if value == None:
        return 0
    else:
        return value

# Fill data as per requirement
company = 'Bajaj_Allianz'
financial_yrs = ['2017_18','2018_19','2019_20','2020_21']
quarters = ['Q1','Q2','Q3','Q4']


odir_path = r"C:\Users\n179950\Munich Re\MR India NL - Documents\Agro\6. Miscellaneous\Individual's Folders\Anil\project_public_disclosure\CSV"
if not os.path.exists(os.path.join(odir_path,company)):
    os.makedirs(os.path.join(odir_path,company))
    
ofile = os.path.join(odir_path,company,'All_year_NL-4.csv')
print(ofile)


Premium_direct = []
RI_accepted = []
RI_ceded =[]
Net_written_premium = []
Earned_premium =[]

all_yr = []
all_qtr =[]
all_cmpny =[]
all_LOB =[]


for financial_yr in financial_yrs:
    for quarter in quarters:
        LOB = ['Fire', 'Marine Cargo', 'Marine Others', 'Marine Total', 'Motor OD', 'Motor TP', 'Motor Total', 'Liability', 'Engineering', 'PA', 'Health Insurance', 'Crop Insurance', 'Others']
        year = np.full(shape=len(LOB),fill_value=financial_yr)
        qtr = np.full(shape=len(LOB),fill_value=quarter)
        cmpny = np.full(shape=len(LOB),fill_value=company.replace('_',' '))
        
        all_LOB.append(LOB)
        all_yr.append(year)
        all_qtr.append(qtr)
        all_cmpny.append(cmpny)
        #financial_yr = '2019_20'
        #quarter = 'Q4'
        
        dir_path = r"C:\Users\n179950\Munich Re\MR India NL - Documents\Agro\6. Miscellaneous\Individual's Folders\Anil\project_public_disclosure\PDF"
        file = os.path.join(dir_path,company,financial_yr,quarter,'NL-4.pdf')
        print(file)
        
        
        
        tables = camelot.read_pdf(file)#,pages="all")
        print("Total tables extracted:", tables.n)
        
        ##### Read Data from NL4 form
        
        #Premium from direct business
        premium_direct_business_Fire = chk_blank(tables[0].df[1][4].replace(',',''))
        premium_direct_business_Marine_cargo = chk_blank(tables[0].df[2][4].replace(',',''))
        premium_direct_business_Marine_others = chk_blank(tables[0].df[3][4].replace(',',''))
        premium_direct_business_Marine_total = chk_blank(tables[0].df[4][4].replace(',',''))
        premium_direct_business_Motor_OD = chk_blank(tables[0].df[5][4].replace(',',''))
        premium_direct_business_Motor_TP = chk_blank(tables[0].df[6][4].replace(',',''))
        premium_direct_business_Motor_Total = chk_blank(tables[0].df[7][4].replace(',',''))
        premium_direct_worksmens = chk_blank(tables[0].df[8][4].replace(',',''))
        premium_direct_business_Liability = chk_blank(tables[0].df[9][4].replace(',',''))
        premium_direct_business_Engineering = chk_blank(tables[0].df[10][4].replace(',',''))
        premium_direct_business_aviation = chk_blank(tables[0].df[11][4].replace(',',''))
        premium_direct_business_PA = chk_blank(tables[0].df[12][4].replace(',',''))
        premium_direct_business_Health_Insurance = chk_blank(tables[0].df[13][4].replace(',',''))
        premium_direct_business_Credit_Insurance = chk_blank(tables[0].df[14][4].replace(',',''))
        premium_direct_business_Crop_Insurance = chk_blank(tables[0].df[15][4].replace(',',''))
        premium_direct_business_Others = chk_blank(tables[0].df[16][4].replace(',',''))
        
        ### Add others all LoB
        premium_direct_business_Others = ifnone(premium_direct_business_Others) + ifnone(premium_direct_worksmens) \
                               + ifnone(premium_direct_business_aviation) + ifnone(premium_direct_business_Credit_Insurance)

        Premium_direct.append(premium_direct_business_Fire)
        Premium_direct.append(premium_direct_business_Marine_cargo)
        Premium_direct.append(premium_direct_business_Marine_others)
        Premium_direct.append(premium_direct_business_Marine_total)
        Premium_direct.append(premium_direct_business_Motor_OD)
        Premium_direct.append(premium_direct_business_Motor_TP)
        Premium_direct.append(premium_direct_business_Motor_Total)
        Premium_direct.append(premium_direct_business_Liability)
        Premium_direct.append(premium_direct_business_Engineering)
        Premium_direct.append(premium_direct_business_PA)
        Premium_direct.append(premium_direct_business_Health_Insurance)
        Premium_direct.append(premium_direct_business_Crop_Insurance)
        Premium_direct.append(premium_direct_business_Others)
        
        # RI Accepted Premium
        RI_accepted_Fire = chk_blank(tables[0].df[1][6].replace(',',''))
        RI_accepted_Marine_cargo = chk_blank(tables[0].df[2][6].replace(',',''))
        RI_accepted_Marine_others = chk_blank(tables[0].df[3][6].replace(',',''))
        RI_accepted_Marine_total = chk_blank(tables[0].df[4][6].replace(',',''))
        RI_accepted_Motor_OD = chk_blank(tables[0].df[5][6].replace(',',''))
        RI_accepted_Motor_TP = chk_blank(tables[0].df[6][6].replace(',',''))
        RI_accepted_Motor_Total = chk_blank(tables[0].df[7][6].replace(',',''))
        RI_accepted_worksmens = chk_blank(tables[0].df[8][6].replace(',',''))
        RI_accepted_Liability = chk_blank(tables[0].df[9][6].replace(',',''))
        RI_accepted_Engineering = chk_blank(tables[0].df[10][6].replace(',',''))
        RI_accepted_aviation = chk_blank(tables[0].df[11][6].replace(',',''))
        RI_accepted_PA = chk_blank(tables[0].df[12][6].replace(',',''))
        RI_accepted_Health_Insurance = chk_blank(tables[0].df[13][6].replace(',',''))
        RI_accepted_Credit_Insurance = chk_blank(tables[0].df[14][6].replace(',',''))
        RI_accepted_Crop_Insurance = chk_blank(tables[0].df[15][6].replace(',',''))
        RI_accepted_Others = chk_blank(tables[0].df[16][6].replace(',',''))
        
        RI_accepted_Others = ifnone(RI_accepted_Others) + ifnone(RI_accepted_worksmens) \
                   + ifnone(RI_accepted_aviation) + ifnone(RI_accepted_Credit_Insurance)
        
        RI_accepted.append(RI_accepted_Fire)
        RI_accepted.append(RI_accepted_Marine_cargo)
        RI_accepted.append(RI_accepted_Marine_others)
        RI_accepted.append(RI_accepted_Marine_total)
        RI_accepted.append(RI_accepted_Motor_OD)
        RI_accepted.append(RI_accepted_Motor_TP)
        RI_accepted.append(RI_accepted_Motor_Total)
        RI_accepted.append(RI_accepted_Liability)
        RI_accepted.append(RI_accepted_Engineering)
        RI_accepted.append(RI_accepted_PA)
        RI_accepted.append(RI_accepted_Health_Insurance)
        RI_accepted.append(RI_accepted_Crop_Insurance)
        RI_accepted.append(RI_accepted_Others)
        
        # RI Ceded Premium
        
        RI_ceded_Fire = chk_blank(tables[0].df[1][10].replace(',',''))
        RI_ceded_Marine_cargo = chk_blank(tables[0].df[2][10].replace(',',''))
        RI_ceded_Marine_others = chk_blank(tables[0].df[3][10].replace(',',''))
        RI_ceded_Marine_total = chk_blank(tables[0].df[4][10].replace(',',''))
        RI_ceded_Motor_OD = chk_blank(tables[0].df[5][10].replace(',',''))
        RI_ceded_Motor_TP = chk_blank(tables[0].df[6][10].replace(',',''))
        RI_ceded_Motor_Total = chk_blank(tables[0].df[7][10].replace(',',''))
        RI_ceded_worksmens = chk_blank(tables[0].df[8][10].replace(',',''))
        RI_ceded_Liability = chk_blank(tables[0].df[9][10].replace(',',''))
        RI_ceded_Engineering = chk_blank(tables[0].df[10][10].replace(',',''))
        RI_ceded_aviation = chk_blank(tables[0].df[11][10].replace(',',''))
        RI_ceded_PA = chk_blank(tables[0].df[12][10].replace(',',''))
        RI_ceded_Health_Insurance = chk_blank(tables[0].df[13][10].replace(',',''))
        RI_ceded_Credit_Insurance = chk_blank(tables[0].df[14][10].replace(',',''))
        RI_ceded_Crop_Insurance = chk_blank(tables[0].df[15][10].replace(',',''))
        RI_ceded_Others = chk_blank(tables[0].df[16][10].replace(',',''))
        
        RI_ceded_Others = ifnone(RI_ceded_Others) + ifnone(RI_ceded_worksmens) \
                + ifnone(RI_ceded_aviation) + ifnone(RI_ceded_Credit_Insurance)
        
        
        RI_ceded.append(RI_ceded_Fire)
        RI_ceded.append(RI_ceded_Marine_cargo)
        RI_ceded.append(RI_ceded_Marine_others)
        RI_ceded.append(RI_ceded_Marine_total)
        RI_ceded.append(RI_ceded_Motor_OD)
        RI_ceded.append(RI_ceded_Motor_TP)
        RI_ceded.append(RI_ceded_Motor_Total)
        RI_ceded.append(RI_ceded_Liability)
        RI_ceded.append(RI_ceded_Engineering)
        RI_ceded.append(RI_ceded_PA)
        RI_ceded.append(RI_ceded_Health_Insurance)
        RI_ceded.append(RI_ceded_Crop_Insurance)
        RI_ceded.append(RI_ceded_Others)
        
        # Net Written Premium
        
        
        Net_written_Premium_Fire = chk_blank(tables[0].df[1][12].replace(',',''))
        Net_written_Premium_Marine_cargo = chk_blank(tables[0].df[2][12].replace(',',''))
        Net_written_Premium_Marine_others = chk_blank(tables[0].df[3][12].replace(',',''))
        Net_written_Premium_Marine_total = chk_blank(tables[0].df[4][12].replace(',',''))
        Net_written_Premium_Motor_OD = chk_blank(tables[0].df[5][12].replace(',',''))
        Net_written_Premium_Motor_TP = chk_blank(tables[0].df[6][12].replace(',',''))
        Net_written_Premium_Motor_Total = chk_blank(tables[0].df[7][12].replace(',',''))
        Net_written_Premium_worksmens = chk_blank(tables[0].df[8][12].replace(',',''))
        Net_written_Premium_Liability = chk_blank(tables[0].df[9][12].replace(',',''))
        Net_written_Premium_Engineering = chk_blank(tables[0].df[10][12].replace(',',''))
        Net_written_Premium_aviation = chk_blank(tables[0].df[11][12].replace(',',''))
        Net_written_Premium_PA = chk_blank(tables[0].df[12][12].replace(',',''))
        Net_written_Premium_Health_Insurance = chk_blank(tables[0].df[13][12].replace(',',''))
        Net_written_Premium_Credit_Insurance = chk_blank(tables[0].df[14][12].replace(',',''))
        Net_written_Premium_Crop_Insurance = chk_blank(tables[0].df[15][12].replace(',',''))
        Net_written_Premium_Others = chk_blank(tables[0].df[16][12].replace(',',''))
        
        Net_written_Premium_Others = ifnone(Net_written_Premium_Others) + ifnone(Net_written_Premium_worksmens) \
                           + ifnone(Net_written_Premium_aviation) + ifnone(Net_written_Premium_Credit_Insurance)
        
        
        Net_written_premium.append(Net_written_Premium_Fire)
        Net_written_premium.append(Net_written_Premium_Marine_cargo)
        Net_written_premium.append(Net_written_Premium_Marine_others)
        Net_written_premium.append(Net_written_Premium_Marine_total)
        Net_written_premium.append(Net_written_Premium_Motor_OD)
        Net_written_premium.append(Net_written_Premium_Motor_TP)
        Net_written_premium.append(Net_written_Premium_Motor_Total)
        Net_written_premium.append(Net_written_Premium_Liability)
        Net_written_premium.append(Net_written_Premium_Engineering)
        Net_written_premium.append(Net_written_Premium_PA)
        Net_written_premium.append(Net_written_Premium_Health_Insurance)
        Net_written_premium.append(Net_written_Premium_Crop_Insurance)
        Net_written_premium.append(Net_written_Premium_Others)
        
        #Premium Earned
        
        
        
        Earned_Premium_Fire = chk_blank(tables[0].df[1][15].replace(',',''))
        Earned_Premium_Marine_cargo = chk_blank(tables[0].df[2][15].replace(',',''))
        Earned_Premium_Marine_others = chk_blank(tables[0].df[3][15].replace(',',''))
        Earned_Premium_Marine_total = chk_blank(tables[0].df[4][15].replace(',',''))
        Earned_Premium_Motor_OD = chk_blank(tables[0].df[5][15].replace(',',''))
        Earned_Premium_Motor_TP = chk_blank(tables[0].df[6][15].replace(',',''))
        Earned_Premium_Motor_Total = chk_blank(tables[0].df[7][15].replace(',',''))
        Earned_Premium_worksmens = chk_blank(tables[0].df[8][15].replace(',',''))
        Earned_Premium_Liability = chk_blank(tables[0].df[9][15].replace(',',''))
        Earned_Premium_Engineering = chk_blank(tables[0].df[10][15].replace(',',''))
        Earned_Premium_aviation = chk_blank(tables[0].df[11][15].replace(',',''))
        Earned_Premium_PA = chk_blank(tables[0].df[12][15].replace(',',''))
        Earned_Premium_Health_Insurance = chk_blank(tables[0].df[13][15].replace(',',''))
        Earned_Premium_Credit_Insurance = chk_blank(tables[0].df[14][15].replace(',',''))
        Earned_Premium_Crop_Insurance = chk_blank(tables[0].df[15][15].replace(',',''))
        Earned_Premium_Others = chk_blank(tables[0].df[16][15].replace(',',''))
        
        Earned_Premium_Others = ifnone(Earned_Premium_Others) + ifnone(Earned_Premium_worksmens) \
                      + ifnone(Earned_Premium_aviation) + ifnone(Earned_Premium_Credit_Insurance)
        
        Earned_premium.append(Earned_Premium_Fire)
        Earned_premium.append(Earned_Premium_Marine_cargo)
        Earned_premium.append(Earned_Premium_Marine_others)
        Earned_premium.append(Earned_Premium_Marine_total)
        Earned_premium.append(Earned_Premium_Motor_OD)
        Earned_premium.append(Earned_Premium_Motor_TP)
        Earned_premium.append(Earned_Premium_Motor_Total)
        Earned_premium.append(Earned_Premium_Liability)
        Earned_premium.append(Earned_Premium_Engineering)
        Earned_premium.append(Earned_Premium_PA)
        Earned_premium.append(Earned_Premium_Health_Insurance)
        Earned_premium.append(Earned_Premium_Crop_Insurance)
        Earned_premium.append(Earned_Premium_Others)

data = {'Year': np.asarray(all_yr).flatten(),
         'Qtr': np.asarray(all_qtr).flatten(),
         'Company': np.asarray(all_cmpny).flatten(),
         'LOB': np.asarray(all_LOB).flatten(),
         'Premium Direct':Premium_direct,
         'RI Accepted':RI_accepted,
         'RI Ceded':RI_ceded,
         'Net Written Premium':Net_written_premium,
         'Earned Premium':Earned_premium}

df_data = pd.DataFrame(data)

df_data.to_csv(ofile,index=False)
print("Completed")
```
