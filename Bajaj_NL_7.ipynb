```python
def chk_blank(value):
    if value =='-' and len(value)==1:
        return None
    elif len(value)==0:
        return None
    elif '(' in value:
        return int(value.replace('-','').replace('(','').replace(')','').replace(' ','')) *-1
    else:
        return int(value.replace('-','').replace(' ',''))
    
def ifnone(value):
    if value == None:
        return 0
    else:
        return value
```

```python
tables = camelot.read_pdf(file)#,pages="all")
print("Total tables extracted:", tables.n)
```

```python
##### Read Data from NL7 form

#Direct Commission
Total = []

Total_business_Fire = chk_blank(tables[0].df[2][34].replace(',',''))
Total_business_Marine_cargo = chk_blank(tables[0].df[3][34].replace(',',''))
Total_business_Marine_others = chk_blank(tables[0].df[4][34].replace(',',''))
Total_business_Marine_total = chk_blank(tables[0].df[5][34].replace(',',''))
Total_business_Motor_OD = chk_blank(tables[0].df[6][34].replace(',',''))
Total_business_Motor_TP = chk_blank(tables[0].df[7][34].replace(',',''))
Total_business_Motor_Total = chk_blank(tables[0].df[8][34].replace(',',''))
Total_business_worksmens = chk_blank(tables[0].df[9][34].replace(',',''))
Total_business_Liability = chk_blank(tables[0].df[10][34].replace(',',''))
Total_business_Engineering = chk_blank(tables[0].df[11][34].replace(',',''))
Total_business_aviation = chk_blank(tables[0].df[12][34].replace(',',''))
Total_business_PA = chk_blank(tables[0].df[13][34].replace(',',''))
Total_business_Health_Insurance = chk_blank(tables[0].df[14][34].replace(',',''))
Total_business_Credit_Insurance = chk_blank(tables[0].df[15][34].replace(',',''))
Total_business_Crop_Insurance = chk_blank(tables[0].df[16][34].replace(',',''))
Total_business_Others = chk_blank(tables[0].df[17][34].replace(',',''))


### Add others all LoB
Total_business_Others = ifnone(Total_business_Others) + ifnone(Total_business_worksmens) \
                               + ifnone(Total_business_aviation) + ifnone(Total_business_Credit_Insurance)

Total.append(Total_business_Fire)
Total.append(Total_business_Marine_cargo)
Total.append(Total_business_Marine_others)
Total.append(Total_business_Marine_total)
Total.append(Total_business_Motor_OD)
Total.append(Total_business_Motor_TP)
Total.append(Total_business_Motor_Total)
Total.append(Total_business_Liability)
Total.append(Total_business_Engineering)
Total.append(Total_business_PA)
Total.append(Total_business_Health_Insurance)
Total.append(Total_business_Crop_Insurance)
Total.append(Total_business_Others)
```
